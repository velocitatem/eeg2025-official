```python
from eegdash import EEGChallengeDataset
from eeg_events import EEGEventExtractor, EEGEventSequenceDataset, EEGEventTransformer
from pathlib import Path
```

```python
data_dir = Path("data")
data_dir.mkdir(parents=True, exist_ok=True)

d = EEGChallengeDataset(
    task="RestingState",
    release="R1", 
    cache_dir=data_dir,
    mini=True
)
```

```python
A=d.datasets[0].raw
```

```python
# processing a single individuals data into raw data, then we pass each person.task recording through event extraction to build a dataset
df=A.to_data_frame() # columns: time, E1, ... E128, Cz
# turn to np array Nx128 - exclude Cz
df = df.drop(columns=["Cz", "time"])
df = df.to_numpy()
df.shape # (40400, 128)
```

```python
# extraction step with eeg_events.py

extractor = EEGEventExtractor()
events = extractor.extract_events(d) # list of EEGEvent objects
print(events)
```
